<h4> DATA CLEANING

```python
df.fillna(value = 0, inplace = True)
df.total_vaccinations = df.total_vaccinations.astype(int)
df.people_vaccinated = df.people_vaccinated.astype(int)
df.people_fully_vaccinated = df.people_fully_vaccinated.astype(int)
df.daily_vaccinations_raw = df.daily_vaccinations_raw.astype(int)
df.total_vaccinations_per_hundred = df.total_vaccinations_per_hundred.astype(int)
df.people_vaccinated_per_hundred = df.people_vaccinated_per_hundred.astype(int)
df.people_fully_vaccinated_per_hundred = df.people_fully_vaccinated_per_hundred.astype(int)
df.daily_vaccinations_per_million = df.daily_vaccinations_per_million.astype(int)
```

```python
df.drop(['source_website'], axis = 1, inplace = True)
```

```python
date = df.date.str.split('-', expand=True)
date
```

```python
df['year'] = date[0]
df['month'] = date[1]
df['day'] = date[2]

df.year = pd.to_numeric(df.year)
df.month = pd.to_numeric(df.month)
df.day = pd.to_numeric(df.day)
```

```python
df.date = pd.to_datetime(df.date)
df.head()
```

<h4>EXPLORATORY DATA ANALYSIS

Top 10 countries on vaccination drive?

```python
Tcountry = df.groupby(['country']).sum().reset_index()
Tcountry
```

```python
TDA = df.groupby(['country']).sum()['people_vaccinated'].reset_index()
```

```python
TDA = TDA.sort_values('people_vaccinated', ascending = False).head(10)
```

```python
TDA
```

```python
plt.figure(figsize=(16,7))
plt.bar(TDA.country,TDA.people_vaccinated)
plt.xticks(rotation= 'vertical');
plt.xlabel('Countries')
plt.ylabel('Total People Vaccinated');
```

Type of Vaccination?

```python
TYV = df.groupby(['vaccines']).sum()['daily_vaccinations_per_million'].reset_index()
```

```python
TYV = TYV.sort_values('daily_vaccinations_per_million', ascending = False)
```

```python
TYV
```

```python
plt.figure(figsize=(12,6))
plt.bar(TYV.vaccines,TYV.daily_vaccinations_per_million);
plt.xticks(rotation = '280');
```
